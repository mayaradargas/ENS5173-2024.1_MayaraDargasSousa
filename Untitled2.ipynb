```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Definindo o caminho do arquivo CSV
file_path = r"C:\ENS5173_MayaraDargas\ENS5173-2024.1_MayaraDargasSousa\inputs\MAX\max_pm10_6.csv"

# Lendo o arquivo CSV
max_pm10_6 = pd.read_csv(file_path)

# Contando o número total de violações (número de linhas)
total_violations = len(max_pm10_6)
print(f"Número total de violações: {total_violations}")

# Contando quantas vezes uma mesma coordenada (x, y) tem violação
violations_per_coordinate = max_pm10_6.groupby(['x', 'y']).size().reset_index(name='violations')

# Exportando a planilha com o número de violações por coordenada para um novo arquivo CSV
output_file_path = r"C:\ENS5173_MayaraDargas\ENS5173-2024.1_MayaraDargasSousa\outputs\violations_per_coordinate_PM10_6.csv"
violations_per_coordinate.to_csv(output_file_path, index=False)
print(f"Planilha exportada para {output_file_path}")

import os
import pandas as pd
import matplotlib.pyplot as plt

# Definindo o caminho do arquivo CSV
file_path = r"C:\ENS5173_MayaraDargas\ENS5173-2024.1_MayaraDargasSousa\inputs\MAX\max_pm10_6.csv"

# Lendo o arquivo CSV
max_pm10_6 = pd.read_csv(file_path)

# Contando quantas vezes uma mesma coordenada (x, y) tem violação
violations_per_coordinate = max_pm10_6.groupby(['x', 'y']).size().reset_index(name='violations')

# Criando um gráfico de dispersão para mostrar a frequência das violações por coordenadas
plt.figure(figsize=(10, 6))

# Ajuste do tamanho dos pontos e mapeamento de cores
scatter = plt.scatter(violations_per_coordinate['x'], violations_per_coordinate['y'], 
                      s=100,  # Tamanho dos pontos
                      c=violations_per_coordinate['violations'],  # Mapeamento de cores baseado em 'violations'
                      cmap='viridis',  # Esquema de cores
                      vmin=1, vmax=20,  # Intervalo de cores de 1 a 20
                      alpha=0.6, edgecolors='w')

# Adicionando uma barra de cor
plt.colorbar(scatter, label='Frequência de Violações')

plt.xlabel('Coordenada X')
plt.ylabel('Coordenada Y')
plt.title('Frequência das Violações por Coordenadas')

# Salvando o gráfico
output_graph_path = r"C:\ENS5173_MayaraDargas\ENS5173-2024.1_MayaraDargasSousa\outputs\violations_frequency_plot_PM10_6.png"
plt.savefig(output_graph_path, dpi=300)

plt.show()

print(f"Gráfico salvo em {output_graph_path}")
```
