```python
import numpy as np #importa a biblioteca Numpy com apelido np
#Criando vetor de dados aleatórios
data = np.random.rand(1000,2) #Cria uma matriz bidimensional (ou vetor de dados) chamada data com 1000 linhas e 2 colunas preenchidas com valores aleatórios entre 0 e 1.

# Verificando o tamanho do vetor
print(data.shape) #diz o tamanho da matriz data
print(np.size(data)) #Imprime o número total de elementos na matriz data, que é a multiplicação do número de linhas pelo número de colunas, ou seja, 1000 * 2 = 2000.

# Print de uma amostra dos dados
print(data[0:10]) #A notação [0:10] seleciona as primeiras 10 linhas da matriz. 
print(data[0:10].shape)
```

```python
# Usando Monte-Carlo - Método Monte Carlo gera uma amostra de dados dentro de um intervalo específico
dataMax = 1000 # Define os limites superior e inferior do intervalo desejado para os dados.
dataMin = 10
data = (dataMax - dataMin)*data + dataMin #Aplica o método de Monte Carlo para transformar uma amostra de dados aleatórios entre 0 e 1 (gerada anteriormente) em uma amostra dentro do intervalo desejado. O método de Monte Carlo é comumente usado para simulações e cálculos numéricos. Aqui, está sendo usado para escalar e transladar os valores aleatórios gerados anteriormente para o intervalo desejado.
print(data[0:10])
```

```python
import matplotlib.pyplot as plt #

fig, ax = plt.subplots(3) #Isso diz ao Python para criar uma área onde podemos desenhar nossos gráficos. O número entre parênteses, neste caso 3, indica quantos gráficos queremos criar. Imagine que queremos criar uma grande tela onde podemos desenhar diferentes gráficos. Aqui, estamos dizendo para o programa preparar uma grande tela com 3 áreas onde podemos desenhar nossos gráficos. Essas áreas são chamadas de subplots. Aqui estamos dizendo ao Python para armazenar essa área onde podemos desenhar nossos gráficos em duas variáveis chamadas fig e ax. A variável fig é como uma folha de papel em branco onde podemos desenhar, enquanto ax é como uma caneta que usaremos para desenhar. Mas, neste caso, ax é mais como uma coleção de canetas, uma para cada uma das áreas que pedimos para criar.
ax[0].plot(data) #aqui, estamos plotando todos os dados em data no primeiro subplot (ax[0]). Se data é uma matriz bidimensional, isso representará a plotagem de ambas as colunas contra o índice das linhas.
ax[1].plot(data[:,0])#Aqui, estamos plotando apenas a primeira coluna de data no segundo subplot (ax[1]). O operador [:,0] seleciona todas as linhas da primeira coluna de data.
ax[2].plot(data[:,1])#Aqui, estamos plotando apenas a segunda coluna de data no terceiro subplot (ax[2]). O operador [:,1] seleciona todas as linhas da segunda coluna de data.
print(type(data)) #Esta linha imprime o tipo de objeto da variável "data"
```

```python
# Criando um dado 2D 
data3d = np.sin(np.random.rand(1000,1000)) #Nesta linha, estamos criando uma matriz 2D de 1000 por 1000 de números aleatórios entre 0 e 1 usando np.random.rand(1000, 1000). Em seguida, estamos aplicando a função seno (np.sin()) a todos esses números. Portanto, estamos calculando o seno de cada número individual na matriz, o que resulta em uma nova matriz com os senos dos números aleatórios.
print(data3d)
```

```python
 #Plotando um dado 3D
fig2, ax2 = plt.subplots() #Aqui, estamos criando uma nova figura (fig2) e um conjunto de eixos (ax2) utilizando a função plt.subplots(). Esta função 
ax2.pcolor(data3d) # Esta linha desenha um gráfico de calor (heatmap) na área de plotagem que criamos anteriormente. A função pcolor() é usada para criar um gráfico de cores, onde os valores na matriz data3d determinam as cores exibidas. Cada célula na matriz será colorida de acordo com o valor do seno do número correspondente na matriz data3d. Os valores mais altos serão representados por cores mais claras e os valores mais baixos por cores mais escuras.
```

```python
# Criando vetor temporal - Um vetor temporal, em termos gerais, é uma sequência de valores que representam momentos ou intervalos de tempo. Ele é usado para representar datas, horas e até mesmo frações de segundos. Em muitos casos, um vetor temporal é uma estrutura de dados que permite manipulações e operações específicas com datas e horas.
dtime = np.arange(np.datetime64("2022-01-01"), np.datetime64("2023-01-01"), np.timedelta64(1, "h")) # Aqui, estamos criando um vetor temporal que começa em 1 de janeiro de 2022 e termina em 1 de janeiro de 2023, com intervalos de 1 hora.
print(dtime[1:10])#aqui estamos imprimindo uma parte do vetor temporal, do segundo ao décimo elemento. Isso nos permite ver como os valores do vetor são representados.
print(dtime.shape)#diz quantos elementos estão no vetor
print(type(dtime))
```

## PANDAS

```python
import pandas as pd

# Criando um dataframe
df = pd.DataFrame() #Esta linha cria um DataFrame vazio chamado df. Um DataFrame é uma estrutura de dados bidimensional, semelhante a uma planilha ou tabela de banco de dados, que é usada para armazenar e manipular dados de forma tabula
```

```python
# Monstrando um dataframe
df['datetime'] = dtime #Esta linha cria uma nova coluna no DataFrame df chamada 'datetime'. Ao usar df['datetime'], estamos especificando que queremos adicionar uma nova coluna chamada 'datetime' ao DataFrame df.
```

```python
# Criando uma coluna random dentro do dataframe df
df['random'] = np.random.rand(df.shape[0]) #np.random.rand() é uma função da biblioteca NumPy que gera números aleatórios entre 0 e 1. df.shape[0] é o número de linhas no DataFrame df. Portanto, np.random.rand(df.shape[0]) gera uma lista de números aleatórios com o mesmo número de elementos que o DataFrame df tem de linhas.
```

```python
df['random'][1:10] #Retorna os valores da coluna 'random' nas linhas de índice 1 a 9 (excluindo o índice 10) do DataFrame df. Esses valores correspondem às segundas à décimas linhas da coluna 'random' no DataFrame df.
```

```python
# Indexação
df.iloc[:,0] #retorna todos os elementos da primeira coluna do DataFrame df. Esses elementos são retornados como uma série pandas.
```

```python
# Colocando o datetime como index
df = df.set_index(df['datetime']) #coluna 'datetime' se tornará o índice do DataFrame df. Isso significa que cada linha do DataFrame será identificada por uma data e hora específicas, e não mais por números inteiros sequenciais.
df
```

```python
# Criando colunas com ano, mês e dia -  Código adiciona três novas colunas ao DataFrame df: 'year' (ano), 'month' (mês) e 'day' (dia), obtendo essas informações a partir do índice temporal do DataFrame
df['year'] = df.index.year #cria uma nova coluna chamada 'year' no DataFrame df e atribui a ela os anos correspondentes ao índice temporal do DataFrame. A propriedade .index acessa o índice do DataFrame e .year extrai o ano de cada elemento do índice.
df['month'] = df.index.month
df['day'] = df.index.day
```

```python
df
```

```python
tt = np.arange(0,df.shape[0],1) #cria um vetor tt com uma sequência de números inteiros de 0 até o número de linhas no DataFrame df
print(tt.shape) #printa o tamanho do dataframe
df['sazonal'] = (np.cos(tt/(360*4))) #Cria uma nova coluna chamada 'timeSeries' no DataFrame df, onde cada valor é calculado como 10 vezes o valor da coluna 'sazonal' mais o valor da coluna 'random'. Isso combina a componente sazonal com algum ruído aleatório representado pela coluna 'random'.
df['timeSeries'] = 10*df['sazonal']+df['random']#Cria uma nova coluna chamada 'timeSeries' no DataFrame df, onde cada valor é calculado como 10 vezes o valor da coluna 'sazonal' mais o valor da coluna 'random'.
#df['sazonal'].plot()
df['timeSeries'].plot(color='red') #Plota a coluna 'timeSeries' do DataFrame df, onde os valores são mostrados em vermelho.

#Em contextos de processamento de sinais e análise de dados, "ruído" refere-se a variações ou perturbações aleatórias indesejadas que podem estar presentes nos dados. Esse ruído pode surgir de várias fontes, como interferências elétricas, erros de medição, flutuações ambientais, entre outros.
#O ruído pode afetar a qualidade dos dados e dificultar a identificação de padrões ou tendências subjacentes. Em muitos casos, é importante filtrar ou remover o ruído dos dados para melhorar a precisão das análises.
#Quando falamos sobre adicionar "ruído aleatório" a um conjunto de dados, geralmente estamos introduzindo variações aleatórias que não têm relação com o sinal ou fenômeno que estamos estudando. Isso pode ser feito com o objetivo de simular condições mais realistas ou testar a robustez de algoritmos de análise de dados em cenários mais desafiadores.
```

```python
# Criando uma função
def noiseCreation(path, year): #Define a função 
    """Esta função cria um ruido com sazonalidade.
    Inputs:
        path = asdasdas
        year = asdasdasdas

    Criado por Leonardo
    """
    
    # Criando um dataframe
    df = pd.DataFrame()
    
    # Criando vertor temporal
    dtime = np.arange(np.datetime64(str(year)+"-01-01"), np.datetime64(str(year+1)+"-01-01"), np.timedelta64(1, "h"))
    
    # Monstrando um dataframe
    df['datetime'] = dtime
        
    # Datetime como index do dataframe
    df = df.set_index(df['datetime'])
    
    # Criando colunas com ano, mês e dia
    df['year'] = df.index.year
    df['month'] = df.index.month
    df['day'] = df.index.day
    
    # Crio uma coluna random
    df['random'] = np.random.rand(df.shape[0])
    
    # Adicionar sazonalidade
    tt = np.arange(0,df.shape[0],1)
    print(tt.shape)
    df['sazonal'] = (np.cos(tt/(360*4)))
    df['timeSeries'] = 10*df['sazonal']+df['random']
    #df['sazonal'].plot()
    df['timeSeries'].plot(color='red')
    df.to_csv(path+'/noise_'+str(year)+'.csv')
    
    return df
    #Reproduz o que vimos anterioirmente mas atirbuindo os códigos a uma função e no final retonrnando ela.
    #Imagine que você tem uma tabela de dados, como uma planilha, onde cada linha representa uma entrada de dados e cada coluna representa um tipo específico de informação, como nome, idade, data de nascimento, etc.
#Quando você define uma coluna específica como índice dessa tabela, você está escolhendo essa coluna para ser a "identificação única" de cada linha. É como se você estivesse dizendo: "Esta coluna contém informações que nos permitem encontrar rapidamente uma linha específica na tabela"""
    
```

```python
import os #Isso importa o módulo os, que fornece funcionalidades relacionadas ao sistema 
files = os.listdir(path)
print(files)

allData = []#nicializa uma lista vazia chamada allData, que será usada para armazenar os dados dos arquivos CSV
for file in files:
    if file.endswith('.csv'):
        print(file)
        data = pd.read_csv(path+'/'+file)
        allData.append(data)
```

```python
allDataDf = pd.concat(allData) #Em resumo, essa linha de código combina todos os DataFrames contidos na lista allData em um único DataFrame chamado allDataDf, o que facilita a análise e manipulação dos dados em conjunto.
```
