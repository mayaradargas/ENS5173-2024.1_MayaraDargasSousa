```python
#Criando um matriz numpy
import numpy as np

#Criando uma matriz de zero utilizando np.zeros
a = np.zeros((5,5))
print(a)
print('Indexação numpy')
print(a[0,0])
```

```python
b = np.random.rand(5,5)
print(b)
print('Dimensão da matriz')
print(b.shape)
```

```python
#Criando matriz de valores aleatórios entre um faixa estabelecida
vmin = 1 #valor mínimo
vmax = 100 #valor máximo 

c=vmin+(vmax-vmin)*b
print(c)
```

```python
#Plotasr matriz 
import matplotlib.pyplot as plt 
plt.plot(c.transpose()) 
#se quiser plotar sem transpor matplotlib
#plota normal (linha) compara pontos no mesmo dia, plota tranposta (coluna) compara mesmo ponto no mesmo dia 
```

```python
#Plotando apenas a primeira linha
plt.plot(c[0,:]) #compara pontos 
```

```python
#Plotando apenas a primeira linha - colocando um eixo x
x = np.array(range(0,np.size(c[0,:])))
print(x)
plt.plot(x,c[0,:]) #compara pontos 
plt.title('Gráfico com valores inteiros no eixo x')
plt.xticks(x) #definindo valores de x
```

```python
#Crinaod um gráfico de barras e exportando 
#primeiro criar nova figura para não sobrecrever
fig, ax = plt.subplots(3,1)
#ax - eixos 
```

```python
#Criando um gráfico de barras e exportando 
#primeiro criar nova figura para não sobrecrever
fig, ax = plt.subplots(3,1)
#ax - eixos 

ax[0].bar(x,c[:,0])
ax[1].bar(x,c[:,1])
ax[2].bar(x,c[:,2])

#exportar a figura 
fig.savefig('Figure01.png')
```

```python
#Criando um gráfico de barras e exportando para local predefinido  
#primeiro criar nova figura para não sobrecrever
fig, ax = plt.subplots(3,1)
#ax - eixos 

ax[0].bar(x,c[:,0])
ax[1].bar(x,c[:,1])
ax[2].bar(x,c[:,2])

#exportar a figura 
import os
rootFolder = os.getcwd()
figFolder = rootFolder.replace('\\','/')+'/figures/'
print(figFolder)
print(rootFolder)
fig.savefig(figFolder+'Figure01.png')
```
